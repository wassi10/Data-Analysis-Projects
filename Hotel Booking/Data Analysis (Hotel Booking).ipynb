# Data Analysis and Visualizations

```python
# in total dataset revervation status

canceled_perc = df['is_canceled'].value_counts(normalize = True)
print(canceled_perc)

plt.figure(figsize = (4,4))
plt.title("Reservation Status")
plt.bar(['Not Canceled', 'canceled'],df['is_canceled'].value_counts(), edgecolor = 'red', width = 0.8)
```

```python
plt.figure(figsize = (6,4))
axl = sns.countplot( x = 'hotel', hue = 'is_canceled', data = df, palette = 'Greens')
legend_labels,_ = axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1,1))
plt.title("Reservation Status in different hotels", size = 18)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not cancled', 'canceled'])
plt.show()
```

```python
# To show the canceled percentage of resort hotel
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'] .value_counts(normalize = True)
```

```python
# To show the canceled percentage of city hotel
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'] .value_counts(normalize = True)
```

```python
# price to resort hotel and city hotel different 
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily rate in city hotel and resort hotel', size = 22)
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
#Which months have more reservations and more cancellations.
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,7))
axl = sns.countplot( x = 'month', hue = 'is_canceled', data = df, palette = 'bright')
legend_labels,_ = axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1,1))
plt.title("Reservation Status per month", size = 18)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not cancled', 'canceled'])
plt.show()



# January has highest cancellation rate
# In january and december has low reservation 
# August has high reservation and low cancellations 
```

```python
# Was the reservation cost lower in August for which cancellations were lower?
                                  #OR
# was the cost higher in January which resulted in higher cancellations?

# Let's calculate average daily rate for each month
plt.figure(figsize = (14, 7))
plt.title("ADR per month", fontsize = 20)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
# Which country has the highest cancellation rate? I'll do for top 10 country
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8, 8))
plt.title("Top 10 countries with reservation canceled", fontsize = 20)
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
```
